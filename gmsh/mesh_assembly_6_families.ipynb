```python
import numpy as np
import gmsh 
import sys
```

### Family Mapping 1/4 assembly

```python
C=[(0,0)]
E=[(1,0),(0,1),(2,0),(0,2),(3,0),(0,3),(4,0),(0,4),(5,0),(0,5),(6,0),(0,6),(7,0),(0,7),(8,0),(0,8)]
N=[(1,1),(2,1),(1,2),(3,1),(1,3)]
D=[(1,4),(4,1),(1,5),(6,1),(1,6),(7,1),(1,7),(2,2),(7,3),(3,7),(4,4),(6,4),(4,6),(7,4),(4,7),(6,6),(7,6),(6,7),(7,7),(6,3),(3,6)]
F=[(5,1),(1,5),(8,1),(1,8),(3,2),(2,3),(2,6),(6,2),(2,7),(7,2),(5,3),(3,5),(8,3),(3,8),(5,4),(4,5),(8,4),(4,8),(6,5),(5,6),(7,5),(5,7),(8,6),(6,8),(8,7),(7,8)]
FD=[(4,2),(2,4),(4,3),(3,4)]
GT=[(5,2),(8,2),(3,3),(2,5),(5,5),(8,5),(2,8),(5,8),[8,8]]
# IT=[8,8]
```

### Family Mapping full assembly

```python
E_left=[]
E_full=[]
for X in E:
    E_left.append((X[0],X[1]))
    E_left.append((X[0],16-X[1]))
for X in E_left:
    E_full.append((X[0],X[1]))
    E_full.append((16-X[0],X[1]))
E_full = list(set(E_full))

C_left=[]
C_full=[]
for X in C:
    C_left.append((X[0],X[1]))
    C_left.append((X[0],16-X[1]))
for X in C_left:
    C_full.append((X[0],X[1]))
    C_full.append((16-X[0],X[1]))
C_full = list(set(C_full))

    
N_left=[]
N_full=[]
for X in N:
    N_left.append((X[0],X[1]))
    N_left.append((X[0],16-X[1]))
for X in N_left:
    N_full.append((X[0],X[1]))
    N_full.append((16-X[0],X[1]))
N_full = list(set(N_full))


D_left=[]
D_full=[]
for X in D:
    D_left.append((X[0],X[1]))
    D_left.append((X[0],16-X[1]))
for X in D_left:
    D_full.append((X[0],X[1]))
    D_full.append((16-X[0],X[1]))   
D_full = list(set(D_full))


F_left=[]
F_full=[]
for X in F:
    F_left.append((X[0],X[1]))
    F_left.append((X[0],16-X[1]))
for X in F_left:
    F_full.append((X[0],X[1]))
    F_full.append((16-X[0],X[1]))    
F_full = list(set(F_full))

FD_left=[]
FD_full=[]
for X in FD:
    FD_left.append((X[0],X[1]))
    FD_left.append((X[0],16-X[1]))
for X in FD_left:
    FD_full.append((X[0],X[1]))
    FD_full.append((16-X[0],X[1]))    
FD_full = list(set(FD_full))


GT_left=[]
GT_full=[]
for X in GT:
    GT_left.append((X[0],X[1]))
    GT_left.append((X[0],16-X[1]))
for X in GT_left:
    GT_full.append((X[0],X[1]))
    GT_full.append((16-X[0],X[1]))    
GT_full = list(set(GT_full))
```

### Create Mesh

```python
gmsh.initialize(sys.argv)
gmsh.model.add("Assembly")

dim = 2
h = 1


pitch = 1.26
r_clad_outer = 0.475
r_fuel = 0.4096
r_clad_inner = 0.418
r_clad = 0.475
r_guide_inner=0.561
r_guide_outer=0.602
r_instru_inner=0.559
r_instru_outer=0.605
dr=2e-1

tag=1
moderator = gmsh.model.occ.addRectangle(-dr/2,dr/2,0,17*pitch+dr,-17*pitch-dr,tag=tag)
tag+=1

C_tag_fuel=[]
E_tag_fuel=[]
N_tag_fuel=[]
D_tag_fuel=[]
F_tag_fuel=[]
FD_tag_fuel=[]
GT_tag_water=[]
IT_tag_water=[]
C_tag_clad=[]
E_tag_clad=[]
N_tag_clad=[]
D_tag_clad=[]
F_tag_clad=[]
FD_tag_clad=[]
GT_tag_clad=[]
IT_tag_clad=[]
C_tag_moderator=[]
E_tag_moderator=[]
N_tag_moderator=[]
D_tag_moderator=[]
F_tag_moderator=[]
FD_tag_moderator=[]
GT_tag_moderator=[]
IT_tag_moderator=[]
C_tag_gap=[]
E_tag_gap=[]
N_tag_gap=[]
D_tag_gap=[]
F_tag_gap=[]
FD_tag_gap=[]
moderator_tag=[1]

pos=[-pitch/2,pitch/2]
for x in range(17):
    for y in range(17):
        X=(x,y)
        pos[0]=x*pitch+pitch/2
        pos[1]=y*(-pitch)-pitch/2        
        if X in C_full:
            moderator_local = gmsh.model.occ.addRectangle(pos[0]-pitch/2,pos[1]+pitch/2,0,pitch,-pitch,tag=tag)
            C_tag_moderator.append(tag)
            tag += 1
            clad = gmsh.model.occ.addDisk(pos[0],pos[1],0,r_clad_outer,r_clad_outer,tag)
            C_tag_clad.append(tag)
            tag += 1
            gap = gmsh.model.occ.addDisk(pos[0],pos[1],0,r_clad_inner,r_clad_inner,tag)
            C_tag_gap.append(tag)
            tag += 1            
            fuel = gmsh.model.occ.addDisk(pos[0],pos[1],0,r_fuel,r_fuel,tag)
            C_tag_fuel.append(tag)
            tag += 1
            gmsh.model.occ.cut([(dim,moderator)],
                    [(dim,moderator_local)],
                    removeTool=False)
            gmsh.model.occ.cut([(dim,moderator_local)],
                    [(dim,clad)],
                    removeTool=False)
            gmsh.model.occ.cut([(dim,clad)],
                    [(dim,gap)],
                    removeTool=False)            
            gmsh.model.occ.cut([(dim,gap)],
                    [(dim,fuel)],
                    removeTool=False)
        if X in E_full:
            moderator_local = gmsh.model.occ.addRectangle(pos[0]-pitch/2,pos[1]+pitch/2,0,pitch,-pitch,tag=tag)
            E_tag_moderator.append(tag)
            tag += 1
            clad = gmsh.model.occ.addDisk(pos[0],pos[1],0,r_clad_outer,r_clad_outer,tag)
            E_tag_clad.append(tag)
            tag += 1
            gap = gmsh.model.occ.addDisk(pos[0],pos[1],0,r_clad_inner,r_clad_inner,tag)
            E_tag_gap.append(tag)
            tag += 1            
            fuel = gmsh.model.occ.addDisk(pos[0],pos[1],0,r_fuel,r_fuel,tag)
            E_tag_fuel.append(tag)
            tag += 1
            gmsh.model.occ.cut([(dim,moderator)],
                    [(dim,moderator_local)],
                    removeTool=False)
            gmsh.model.occ.cut([(dim,moderator_local)],
                    [(dim,clad)],
                    removeTool=False)
            gmsh.model.occ.cut([(dim,clad)],
                    [(dim,gap)],
                    removeTool=False)            
            gmsh.model.occ.cut([(dim,gap)],
                    [(dim,fuel)],
                    removeTool=False)              
        if X in N_full:                
            moderator_local = gmsh.model.occ.addRectangle(pos[0]-pitch/2,pos[1]+pitch/2,0,pitch,-pitch,tag=tag)
            N_tag_moderator.append(tag)
            tag += 1
            clad = gmsh.model.occ.addDisk(pos[0],pos[1],0,r_clad_outer,r_clad_outer,tag)
            N_tag_clad.append(tag)
            tag += 1
            gap = gmsh.model.occ.addDisk(pos[0],pos[1],0,r_clad_inner,r_clad_inner,tag)
            N_tag_gap.append(tag)
            tag += 1            
            fuel = gmsh.model.occ.addDisk(pos[0],pos[1],0,r_fuel,r_fuel,tag)
            N_tag_fuel.append(tag)
            tag += 1
            gmsh.model.occ.cut([(dim,moderator)],
                    [(dim,moderator_local)],
                    removeTool=False)
            gmsh.model.occ.cut([(dim,moderator_local)],
                    [(dim,clad)],
                    removeTool=False)
            gmsh.model.occ.cut([(dim,clad)],
                    [(dim,gap)],
                    removeTool=False)            
            gmsh.model.occ.cut([(dim,gap)],
                    [(dim,fuel)],
                    removeTool=False)
        if X in D_full:
            moderator_local = gmsh.model.occ.addRectangle(pos[0]-pitch/2,pos[1]+pitch/2,0,pitch,-pitch,tag=tag)
            D_tag_moderator.append(tag)
            tag += 1
            clad = gmsh.model.occ.addDisk(pos[0],pos[1],0,r_clad_outer,r_clad_outer,tag)
            D_tag_clad.append(tag)
            tag += 1
            gap = gmsh.model.occ.addDisk(pos[0],pos[1],0,r_clad_inner,r_clad_inner,tag)
            D_tag_gap.append(tag)
            tag += 1            
            fuel = gmsh.model.occ.addDisk(pos[0],pos[1],0,r_fuel,r_fuel,tag)
            D_tag_fuel.append(tag)
            tag += 1
            gmsh.model.occ.cut([(dim,moderator)],
                    [(dim,moderator_local)],
                    removeTool=False)
            gmsh.model.occ.cut([(dim,moderator_local)],
                    [(dim,clad)],
                    removeTool=False)
            gmsh.model.occ.cut([(dim,clad)],
                    [(dim,gap)],
                    removeTool=False)            
            gmsh.model.occ.cut([(dim,gap)],
                    [(dim,fuel)],
                    removeTool=False)
        if X in F_full:
            moderator_local = gmsh.model.occ.addRectangle(pos[0]-pitch/2,pos[1]+pitch/2,0,pitch,-pitch,tag=tag)
            F_tag_moderator.append(tag)
            tag += 1
            clad = gmsh.model.occ.addDisk(pos[0],pos[1],0,r_clad_outer,r_clad_outer,tag)
            F_tag_clad.append(tag)
            tag += 1
            gap = gmsh.model.occ.addDisk(pos[0],pos[1],0,r_clad_inner,r_clad_inner,tag)
            F_tag_gap.append(tag)
            tag += 1            
            fuel = gmsh.model.occ.addDisk(pos[0],pos[1],0,r_fuel,r_fuel,tag)
            F_tag_fuel.append(tag)
            tag += 1
            gmsh.model.occ.cut([(dim,moderator)],
                    [(dim,moderator_local)],
                    removeTool=False)
            gmsh.model.occ.cut([(dim,moderator_local)],
                    [(dim,clad)],
                    removeTool=False)
            gmsh.model.occ.cut([(dim,clad)],
                    [(dim,gap)],
                    removeTool=False)            
            gmsh.model.occ.cut([(dim,gap)],
                    [(dim,fuel)],
                    removeTool=False)
        if X in FD_full:
            moderator_local = gmsh.model.occ.addRectangle(pos[0]-pitch/2,pos[1]+pitch/2,0,pitch,-pitch,tag=tag)
            FD_tag_moderator.append(tag)
            tag += 1
            clad = gmsh.model.occ.addDisk(pos[0],pos[1],0,r_clad_outer,r_clad_outer,tag)
            FD_tag_clad.append(tag)
            tag += 1
            gap = gmsh.model.occ.addDisk(pos[0],pos[1],0,r_clad_inner,r_clad_inner,tag)
            FD_tag_gap.append(tag)
            tag += 1            
            fuel = gmsh.model.occ.addDisk(pos[0],pos[1],0,r_fuel,r_fuel,tag)
            FD_tag_fuel.append(tag)
            tag += 1
            gmsh.model.occ.cut([(dim,moderator)],
                    [(dim,moderator_local)],
                    removeTool=False)
            gmsh.model.occ.cut([(dim,moderator_local)],
                    [(dim,clad)],
                    removeTool=False)
            gmsh.model.occ.cut([(dim,clad)],
                    [(dim,gap)],
                    removeTool=False)            
            gmsh.model.occ.cut([(dim,gap)],
                    [(dim,fuel)],
                    removeTool=False)                                 
        if X in GT_full:
            if X==(8,8):
                IT_moderator_local = gmsh.model.occ.addRectangle(pos[0]-pitch/2,pos[1]+pitch/2,0,pitch,-pitch,tag=tag)
                IT_tag_moderator.append(tag)
                tag += 1            
                IT_tube_clad=gmsh.model.occ.addDisk(pos[0],pos[1],0,r_instru_outer,r_instru_outer,tag)
                IT_tag_clad.append(tag)
                tag += 1
                IT_tube_water = gmsh.model.occ.addDisk(pos[0],pos[1],0,r_instru_inner,r_instru_inner,tag)
                IT_tag_water.append(tag)
                tag += 1
                gmsh.model.occ.cut([(dim,moderator)],
                    [(dim,IT_moderator_local)],
                    removeTool=False)            
                gmsh.model.occ.cut([(dim,IT_moderator_local)],
                    [(dim,IT_tube_clad)],
                    removeTool=False)
                gmsh.model.occ.cut([(dim,IT_tube_clad)],
                    [(dim,IT_tube_water)],
                       removeTool=False) 
            else:
                GT_moderator_local = gmsh.model.occ.addRectangle(pos[0]-pitch/2,pos[1]+pitch/2,0,pitch,-pitch,tag=tag)
                GT_tag_moderator.append(tag)
                tag += 1            
                GT_tube_clad=gmsh.model.occ.addDisk(pos[0],pos[1],0,r_guide_outer,r_guide_outer,tag)
                GT_tag_clad.append(tag)
                tag += 1
                GT_tube_water = gmsh.model.occ.addDisk(pos[0],pos[1],0,r_guide_inner,r_guide_inner,tag)
                GT_tag_water.append(tag)
                tag += 1
                gmsh.model.occ.cut([(dim,moderator)],
                    [(dim,GT_moderator_local)],
                    removeTool=False)            
                gmsh.model.occ.cut([(dim,GT_moderator_local)],
                    [(dim,GT_tube_clad)],
                    removeTool=False)
                gmsh.model.occ.cut([(dim,GT_tube_clad)],
                    [(dim,GT_tube_water)],
                       removeTool=False)
        # if X in IT:
        #     IT_moderator_local = gmsh.model.occ.addRectangle(pos[0]-pitch/2,pos[1]+pitch/2,0,pitch,-pitch,tag=tag)
        #     IT_tag_moderator.append(tag)
        #     tag += 1            
        #     IT_tube_clad=gmsh.model.occ.addDisk(pos[0],pos[1],0,r_instru_outer,r_instru_outer,tag)
        #     IT_tag_clad.append(tag)
        #     tag += 1
        #     IT_tube_water = gmsh.model.occ.addDisk(pos[0],pos[1],0,r_instru_inner,r_instru_inner,tag)
        #     IT_tag_water.append(tag)
        #     tag += 1
        #     gmsh.model.occ.cut([(dim,moderator)],
        #             [(dim,IT_moderator_local)],
        #             removeTool=False)            
        #     gmsh.model.occ.cut([(dim,IT_moderator_local)],
        #             [(dim,IT_tube_clad)],
        #             removeTool=False)
        #     gmsh.model.occ.cut([(dim,IT_tube_clad)],
        #             [(dim,IT_tube_water)],
        #                removeTool=False)            
```

### Cut the water

cutpoint1 = gmsh.model.occ.addPoint(pitch/2,pitch/2,0.0)
cutpoint2 = gmsh.model.occ.addPoint(pitch/2,-pitch/2,0.0)
cutline = gmsh.model.occ.addLine(cutpoint1,cutpoint2)

### Add Mesh Field

gmsh.model.mesh.field.add("Distance", 1)
gmsh.model.mesh.field.setNumbers(1, "CurvesList", [cutline])
gmsh.model.mesh.field.setNumber(1, "Sampling", 100)
gmsh.model.mesh.field.add("Threshold", 2)
gmsh.model.mesh.field.setNumber(2, "InField", 1)
gmsh.model.mesh.field.setNumber(2, "SizeMin", h/2)
gmsh.model.mesh.field.setNumber(2, "SizeMax", h)
gmsh.model.mesh.field.setNumber(2, "DistMin", h*10)
gmsh.model.mesh.field.setNumber(2, "DistMax", h*20)

gmsh.model.mesh.field.add("Min", 4)
gmsh.model.mesh.field.setNumbers(4, "FieldsList", [2])
gmsh.model.mesh.field.setAsBackgroundMesh(4)

### Add Phyisical groups

```python
gmsh.model.occ.synchronize()

def phys_group(group_tag):
    gmsh.model.addPhysicalGroup(dim,tags=group_tag,name=str(group_tag))

group_tag=[moderator_tag,C_tag_fuel,E_tag_fuel,N_tag_fuel,D_tag_fuel,F_tag_fuel,FD_tag_fuel,GT_tag_water,C_tag_clad,E_tag_clad,N_tag_clad,D_tag_clad,F_tag_clad,FD_tag_clad,GT_tag_clad,C_tag_moderator,E_tag_moderator,N_tag_moderator,D_tag_moderator,F_tag_moderator,FD_tag_moderator,GT_tag_moderator,C_tag_gap,E_tag_gap,N_tag_gap,D_tag_gap,F_tag_gap,FD_tag_gap]

for group in group_tag:
    phys_group(group)
```

### Mesh Settings

```python
# gmsh.model.occ.synchronize()
gmsh.option.setNumber("Mesh.MeshSizeExtendFromBoundary", 0)
gmsh.option.setNumber("Mesh.MeshSizeFromPoints", 0)
gmsh.option.setNumber("Mesh.MeshSizeFromCurvature", 0)
gmsh.option.setNumber("Mesh.MeshSizeMax", h)

# Meshing
# Export Mesh to specified version for OpenSn
gmsh.option.setNumber("Mesh.MshFileVersion",2.2)

#Triangle Mesh
gmsh.option.setNumber("Mesh.Algorithm", 6)

#Quads Mesh
# gmsh.model.mesh.setRecombine(dim,1)
# gmsh.option.setNumber("Mesh.Algorithm", 8)
```
