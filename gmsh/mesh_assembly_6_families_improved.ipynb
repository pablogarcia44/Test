```python
import numpy as np
import gmsh 
import sys

dim = 2
h = 1

pitch = 1.26
r_clad = 0.475
r_fuel = 0.418
r_guide_inner=0.561
r_guide_outer=0.602
r_instru_inner=0.559
r_instru_outer=0.605
dr=2e-1
```

```python
def create_mesh_family(family,family_map,tag):
    gmsh.model.occ.synchronize()
    moderator_tag=[]
    clad_tag=[]
    fuel_tag=[]
    pos=[99,99]
    for map_pos in family_map:
        pos[0]=map_pos[0]*pitch+pitch/2
        pos[1]=map_pos[1]*(-pitch)-pitch/2 
        moderator = gmsh.model.occ.addRectangle(pos[0]-pitch/2,pos[1]+pitch/2,0,pitch,-pitch,tag=tag)
        moderator_tag.append(tag)
        tag += 1
        
        clad = gmsh.model.occ.addDisk(pos[0],pos[1],0,r_clad,r_clad,tag)
        clad_tag.append(tag)
        tag += 1
        
        fuel = gmsh.model.occ.addDisk(pos[0],pos[1],0,r_fuel,r_fuel,tag)
        fuel_tag.append(tag)
        tag += 1
        
        gmsh.model.occ.cut([(dim,1)],
                    [(dim,moderator)],
                    removeTool=False)
        gmsh.model.occ.cut([(dim,moderator)],
                    [(dim,clad)],
                    removeTool=False)
        gmsh.model.occ.cut([(dim,clad)],
                    [(dim,fuel)],
                    removeTool=False) 
    return([moderator_tag,clad_tag,fuel_tag,tag])
```
